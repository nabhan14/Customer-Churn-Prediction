```python
import pandas as pd

df = pd.read_excel(
    "raw/Telco_customer_churn.xlsx",
    engine="openpyxl"
)

print(df.head())
print(df.shape)
```

```python
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
```

```python
target = "Churn Label"
```

```python
df[target] = df[target].map({"Yes": 1, "No": 0})
```

```python
df
```

```python
drop_cols = [
    "CustomerID",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Latitude",
    "Longitude",
    "Churn Reason",
    "Churn Score",
    "Churn Value"
]
```

```python
df_model = df.drop(columns=drop_cols)
```

```python
df_model
```

```python
df_model.shape
df_model["Churn Label"].value_counts()
```

```python
df_model.to_csv(
    "processed/Telco_customer_churn.csv",
    index=False
)
```
